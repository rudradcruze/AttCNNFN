```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Data from the table
data = {
    'Feature': ['TF-IDF', 'CV', 'W2V', 'CV + TF-IDF +W2V', 'CV + TF-IDF', 'TF-IDF + W2V', 'CV + W2V'],
    'AUC': [0.9973, 0.9980, 0.9864, 0.9990, 0.9981, 0.9972, 0.9989],
    'ACC': [0.9839, 0.9887, 0.9579, 0.9909, 0.9889, 0.9577, 0.9911],
    'MCC': [0.9678, 0.9774, 0.9177, 0.9817, 0.9779, 0.9183, 0.9822],
    'Sen': [0.9842, 0.9911, 0.9887, 0.9941, 0.9896, 0.9970, 0.9926],
    'Spe': [0.9836, 0.9864, 0.9287, 0.9878, 0.9883, 0.9203, 0.9897],
    'Pre': [0.9828, 0.9858, 0.9295, 0.9873, 0.9877, 0.9224, 0.9892],
    'F1': [0.9835, 0.9884, 0.9582, 0.9907, 0.9887, 0.9583, 0.9909]
}

df = pd.DataFrame(data)

# Plotting the bar graph
bar_width = 0.1
index = np.arange(len(df))

plt.figure(figsize=(10, 5))

# Plot each metric as a separate bar in the bar graph
metrics = ['AUC', 'ACC', 'MCC', 'Sen', 'Spe', 'Pre', 'F1']
colors = ['green', 'orange', 'blue', 'red', 'purple', 'pink', 'magenta']

for i, metric in enumerate(metrics):
    plt.bar(index + i * bar_width, df[metric], bar_width, label=metric, color=colors[i])

# plt.xlabel('Features')
plt.ylabel('Evaluation metrics')
# plt.title('Performance comparison of various features')
plt.xticks(index + bar_width * 3, df['Feature'], rotation=45)
plt.legend(title='Metrics', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

# Show plot
plt.show()
```
